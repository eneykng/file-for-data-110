## Waste treatment vs Income type

```python
anaerob = (hic_mean['anaerobic waste,%'], umc_mean['anaerobic waste,%'], lmc_mean['anaerobic waste,%'], lic_mean['anaerobic waste,%'])
compost = (hic_mean['compost waste,%'], umc_mean['compost waste,%'], lmc_mean['compost waste,%'], lic_mean['compost waste,%'])
controlled_landfill = (hic_mean['controlled landfill waste,%'], umc_mean['controlled landfill waste,%'], lmc_mean['controlled landfill waste,%'], lic_mean['controlled landfill waste,%'])
incineration = (hic_mean['incineration waste,%'], umc_mean['incineration waste,%'], lmc_mean['incineration waste,%'], lic_mean['incineration waste,%'])
unsp_landfill = (hic_mean['unspecified landfill, %'], umc_mean['unspecified landfill, %'], lmc_mean['unspecified landfill, %'], lic_mean['unspecified landfill, %'])
open_damb = (hic_mean['open dumb,%'], umc_mean['open dumb,%'], lmc_mean['open dumb,%'], lic_mean['open dumb,%'])
other = (hic_mean['other treatment, %'], umc_mean['other treatment, %'], lmc_mean['other treatment, %'], lic_mean['other treatment, %'])
recycling = (hic_mean['recycling, %'], umc_mean['recycling, %'], lmc_mean['recycling, %'], lic_mean['recycling, %'])
gas = (hic_mean['gas systen,%'], umc_mean['gas systen,%'], lmc_mean['gas systen,%'], lic_mean['gas systen,%'])
unaccounted = (hic_mean['unaccounted,%'], umc_mean['unaccounted,%'], lmc_mean['unaccounted,%'], lic_mean['unaccounted,%'])
marine = (hic_mean['marine,%'], umc_mean['marine,%'], lmc_mean['marine,%'], lic_mean['marine,%'])
```

```python
treatment_means = anaerob,compost,controlled_landfill,incineration,unsp_landfill,open_damb,other,recycling,gas,unaccounted,marine
```

```python
df_3 = pd.DataFrame(treatment_means, index = ['anaerobic waste,%','compost waste,%','controlled landfill waste,%','incineration waste,%','unspecified landfill, %','open dumb,%','other treatment, %','recycling, %','gas systen,%', 'unaccounted,%','marine,%'],columns = ['HIC', "UMC", 'LMC', 'LIC'])
print(df_3)
```

```python
#This heatmap shows waste treatment for different income types countries
plt.figure(figsize=(10,5))
sns.heatmap(df_3, cmap="flare", linewidths=0.5, annot=False)
plt.title('Waste treatment per income type')
plt.xlabel('Income type')
plt.ylabel('Treatment type')
plt.show()
```

```python
#HIC countries have the most diverse waste treatment
```
